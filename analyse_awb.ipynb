```python
from fractions import  Fraction
import time

import matplotlib.pyplot as plt
import numpy as np
from picamerax import PiCamera
```

```python
with PiCamera(resolution=(320,240)) as camera:
    camera.awb_mode = "auto"
    # camera.exposure_mode = "off"
    camera.analog_gain = 1
    camera.digital_gain = 1
    camera.shutter_speed = int(1e6 * 1 / 4000)
    time.sleep(2)
    output = np.empty((240, 320, 3), dtype=np.uint8)
    print("analog gain", camera.analog_gain)
    print("digital gain", camera.digital_gain)
    camera.capture(output, "rgb")

    print("AWB", camera.awb_mode, camera.awb_gains)
    print("Red Mean", output[:,:,0].mean())
    print("Green Mean", output[:,:,1].mean())
    print("Blue Mean", output[:,:,2].mean())

    plt.figure(figsize=(12,4))
    plt.subplot(121)
    plt.imshow(output)
    plt.subplot(122)
    plt.plot(output[120,:,0], c="r")
    plt.plot(output[120,:,1], c="g")
    plt.plot(output[120,:,2], c="b")
    plt.ylim((0,255))
    plt.grid()
    plt.show()
```
