```python
import pickle

import matplotlib.pyplot as plt
import numpy as np
```

## Full Iteration (RAW)

```python
with open("times_iteration_raw.pkl", "rb") as f:
    data = pickle.load(f)
frame, capture, advance = zip(*data)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(311)
plt.title("Full Iteration (RAW)")
plt.plot(frame, label="All")
plt.plot(capture, label="Capture + DNG")
plt.plot(advance, label="Advance")
plt.ylabel("Time (s)")
plt.xlim([0, None])
plt.ylim([0, None])
plt.legend()
plt.grid()
plt.subplot(312)
plt.title(f"Capture + DNG (mean: {np.mean(capture):.3f}, std: {np.std(capture):.3f})")
plt.plot(capture, c="tab:orange")
m, b = np.polyfit(range(len(data)), capture, 1)
plt.plot(m*np.array(range(len(data)))+b, c="tab:blue", ls="--", label="Linear Fit")
plt.ylabel("Time (s)")
plt.xlim([0, None])
plt.legend()
plt.grid()
plt.subplot(313)
plt.title(f"Advance (mean: {np.mean(advance):.3f}, std: {np.std(advance):.3f})")
plt.plot(advance, c="tab:green")
m, b = np.polyfit(range(len(data)), advance, 1)
plt.plot(m*np.array(range(len(data)))+b, c="tab:red", ls="--", label="Linear Fit")
plt.ylabel("Time (s)")
plt.xlim([0, None])
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

## Capture (RAW)

```python
with open("times_all_capture_raw.pkl", "rb") as f:
    data = pickle.load(f)
full = [sample["tf"] for sample in data]
capture = [sample["tc"] for sample in data]
dng = [sample["tdng"] for sample in data]
write = [sample["tw"] for sample in data]
```

```python
plt.figure(figsize=(10,12.5))
plt.subplot(411)
plt.title("Capture (RAW)")
plt.plot(full, label="All")
plt.plot(capture, label="Capture")
plt.plot(dng, label="DNG Conversion")
plt.plot(write, label="Write")
plt.ylabel("Time (s)")
plt.xlim([0, None])
plt.ylim([0, None])
plt.legend()
plt.grid()
plt.subplot(412)
plt.title(f"Capture (mean: {np.mean(capture):.3f}, std: {np.std(capture):.3f})")
plt.plot(capture, c="tab:orange")
m, b = np.polyfit(range(len(data)), capture, 1)
plt.plot(m*np.array(range(len(data)))+b, c="tab:blue", ls="--", label="Linear Fit")
plt.ylabel("Time (s)")
plt.xlim([0, None])
plt.legend()
plt.grid()
plt.subplot(413)
plt.title(f"DNG Conversion (mean: {np.mean(dng):.3f}, std: {np.std(dng):.3f})")
plt.plot(dng, c="tab:green")
m, b = np.polyfit(range(len(data)), dng, 1)
plt.plot(m*np.array(range(len(data)))+b, c="tab:red", ls="--", label="Linear Fit")
plt.ylabel("Time (s)")
plt.xlim([0, None])
plt.legend()
plt.grid()
plt.subplot(414)
plt.title(f"Write (mean: {np.mean(write):.3f}, std: {np.std(write):.3f})")
plt.plot(write, c="tab:red")
m, b = np.polyfit(range(len(data)), write, 1)
plt.plot(m*np.array(range(len(data)))+b, c="tab:green", ls="--", label="Linear Fit")
plt.ylabel("Time (s)")
plt.xlim([0, None])
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
