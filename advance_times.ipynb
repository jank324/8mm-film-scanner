```python
import pickle

import cv2
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from tqdm import tqdm
```

```python
with open("advance_times_1641761558.4724083.pkl", "rb") as f:
    times = pickle.load(f)
times = np.array(times)

with open("advance_times_1641864830.pkl", "rb") as f:
    times3 = pickle.load(f)
times3 = np.array(times3)
```

```python
plt.figure(figsize=(12,4))
plt.plot(times)
plt.grid()
plt.title(f"Advance Times after Hall Effect Sensor Adjustment (mean: {times.mean():.5f} / std: {times.std():.5f})")
plt.xlabel("Frame")
plt.ylabel("Advance time (s)")
plt.xlim(0, None)
plt.ylim(0, None)
plt.show()

plt.figure(figsize=(12,4))
plt.plot(times3)
plt.grid()
plt.title(f"Advance Times after Hall Effect Sensor Adjustment (mean: {times3.mean():.5f} / std: {times3.std():.5f})")
plt.xlabel("Frame")
plt.ylabel("Advance time (s)")
plt.xlim(0, None)
plt.ylim(0, None)
plt.show()
```

```python
times500 = times[:500]

plt.figure(figsize=(12,4))
plt.plot(times500)
plt.grid()
plt.title(f"Advance Times after Hall Effect Sensor Adjustment (mean: {times500.mean():.5f} / std: {times500.std():.5f})")
plt.xlabel("Frame")
plt.ylabel("Advance time (s)")
plt.xlim(0, None)
plt.ylim(0, None)
plt.show()

times3500 = times3[:500]

plt.figure(figsize=(12,4))
plt.plot(times3500)
plt.grid()
plt.title(f"Advance Times after Hall Effect Sensor Adjustment (mean: {times3500.mean():.5f} / std: {times3500.std():.5f})")
plt.xlabel("Frame")
plt.ylabel("Advance time (s)")
plt.xlim(0, None)
plt.ylim(0, None)
plt.show()
```

```python
sns.displot(times, kde=True, aspect=2)
plt.show()

sns.displot(times3, kde=True, aspect=2)
plt.show()
```

```python
angles3 = times3 * 360 / times3.mean()

plt.figure(figsize=(12,4))
plt.plot(angles3)
plt.grid()
plt.title("Advance Angles after Hall Effect Sensor Adjustment")
plt.xlabel("Frame")
plt.ylabel("Advance Angle (degrees)")
plt.xlim(0, None)
# plt.ylim(0, None)
plt.show()
```

```python
luminosities3 = []
for i in tqdm(range(12000)):
    path = f"/Volumes/PortableSSD/testhalladjust3/frame-{i:05d}.jpg"
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    l = img.mean()
    luminosities3.append(l)
luminosities3 = np.array(luminosities3)

medlum = np.median(luminosities3)
diffs = np.abs(luminosities3 - medlum)

ind = np.argpartition(diffs, -5)[-5:]
print(np.sort(ind))

plt.figure(figsize=(12,4))
plt.plot(luminosities3)
plt.grid()
plt.title("Frame Luminosities after Hall Effect Sensor Adjustment")
plt.xlabel("Frame")
plt.ylabel("Luminosity")
plt.xlim(0, None)
# plt.ylim(0, None)
plt.show()
```
