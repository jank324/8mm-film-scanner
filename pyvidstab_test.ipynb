```python
import matplotlib.pyplot as plt
from vidstab import VidStab
```

```python
stabilizer = VidStab()
stabilizer.gen_transforms(input_path="/Volumes/PortableSSD/pyvidstab_test_original.m4v")
stabilizer.apply_transforms(
    input_path="/Volumes/PortableSSD/pyvidstab_test_original.m4v",
    output_path="/Volumes/PortableSSD/pyvidstab_test_vidstabbed_30_window.m4v"
)
stabilizer.plot_trajectory()
plt.show()
```

```python
stabilizer = VidStab()
stabilizer.gen_transforms(
    input_path="/Volumes/PortableSSD/pyvidstab_test_original.m4v",
    smoothing_window=5
)
stabilizer.apply_transforms(
    input_path="/Volumes/PortableSSD/pyvidstab_test_original.m4v",
    output_path="/Volumes/PortableSSD/pyvidstab_test_vidstabbed_05_window.m4v"
)
stabilizer.plot_trajectory()
plt.show()
```

```python
stabilizer = VidStab()
stabilizer.gen_transforms(
    input_path="/Volumes/PortableSSD/cropend_frame.m4v",
    smoothing_window=5
)
stabilizer.apply_transforms(
    input_path="/Volumes/PortableSSD/cropend_frame.m4v",
    output_path="/Volumes/PortableSSD/cropend_vidstabbed_frame.m4v"
)
stabilizer.plot_trajectory()
plt.show()
stabilizer.plot_transforms()
plt.show()
```

```python
stabilizer = VidStab()
stabilizer.gen_transforms(
    input_path="/Volumes/PortableSSD/cropend_frame.m4v",
    smoothing_window=5
)
stabilizer.transforms[:,0] = 0
stabilizer.apply_transforms(
    input_path="/Volumes/PortableSSD/cropend_frame.m4v",
    output_path="/Volumes/PortableSSD/cropend_vidstabbed_nox_frame.m4v"
)
stabilizer.plot_trajectory()
plt.show()
stabilizer.plot_transforms()
plt.show()
```
